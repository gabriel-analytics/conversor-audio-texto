```python
# Montagem do Google Drive
# Instalação de dependências
# Importação de bibliotecas
# Varredura de arquivos na pasta
# Transcrição com Whisper + timestamps
# Formatação com títulos, separadores e marcação de tempo
# Salvamento do resultado formatado no Google Drive

#  Montar Google Drive para acessar os áudios
from google.colab import drive
drive.mount('/content/drive')

#  Instalar Whisper e ffmpeg
!pip install git+https://github.com/openai/whisper.git
!sudo apt update && sudo apt install ffmpeg

#  Importar bibliotecas necessárias
import os
import whisper

#  Configurar a pasta onde estão os arquivos de áudio no Google Drive
audio_folder = '/content/drive/MyDrive/audios/'

#  Listar e ordenar os arquivos de áudio (apenas .mp3 e .wav)
audio_files = sorted([
    f for f in os.listdir(audio_folder)
    if f.lower().endswith('.mp3') or f.lower().endswith('.wav')
])

print(f" Arquivos encontrados para transcrição: {audio_files}")

#  Carregar o modelo Whisper
# Pode escolher entre: "tiny", "base", "small", "medium", "large"
# Modelos maiores são mais precisos, porém mais lentos e pesados.
model = whisper.load_model("medium")

#  Inicializar lista para armazenar todas as transcrições formatadas
transcricoes = []

#  Loop para processar cada arquivo de áudio
for idx, audio in enumerate(audio_files):
    audio_path = os.path.join(audio_folder, audio)

    print(f"\n🔹 Processando arquivo {idx + 1}/{len(audio_files)}: {audio}")

    #  Transcrever com Whisper, incluindo timestamps
    result = model.transcribe(audio_path, language="pt", word_timestamps=True)

    #  Formatar a transcrição com título e separadores
    transcricao = f"\n\n---\n\n📁 **Arquivo: {audio}**\n\n"

    #  Iterar sobre os segmentos com timestamps
    for segment in result['segments']:
        start = segment['start']
        end = segment['end']
        text = segment['text']

        #  Adicionar ao texto a marcação de tempo e o conteúdo -- comentei esse trecho ##[{start:.2f}s - {end:.2f}s]:
        transcricao += f"{text}\n"

    print(transcricao)

    #  Adicionar transcrição formatada à lista
    transcricoes.append(transcricao)

#  Concatenar todas as transcrições
transcricao_final = '\n'.join(transcricoes)

#  Exibir transcrição final
print("\n Transcrição Final Completa:")
print(transcricao_final)

#  Salvar a transcrição final formatada no Google Drive
output_path = '/content/drive/MyDrive/audios/transcricao_final_formatada.txt'

with open(output_path, 'w') as f:
    f.write(transcricao_final)

print(f"\n Transcrição salva com sucesso em: {output_path}")
#  Resumo do fluxo:
# Drive: monta e lê arquivos automaticamente.
# Whisper: transcreve cada áudio com timestamps.
# Formatação: separa cada áudio com título + linha ---.
# Concatenação: une todas as transcrições.
# Exportação: salva o arquivo consolidado no Google Drive.

```
